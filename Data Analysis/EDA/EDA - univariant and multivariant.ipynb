# Bivariant and Multivariant analysis


```python
fl=sns.load_dataset('flights')
ir=sns.load_dataset('iris')
t=sns.load_dataset('tips')
t.head()
```

## Scatterplot (numerical-numerical)

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x='total_bill',y='tip',data=t, hue='sex',style='smoker',size='size')
```

## Barplot (numerical-categorical)

```python
da
```

```python
sns.barplot(x='Pclass',y='Fare',data=da, hue='Sex')
```

## Boxplot (numerical-categorical)

```python
sns.boxplot(x='Pclass',y='Age',data=da, hue='Sex')
```

## Distplot (numerical-categorical)

```python
sns.distplot(da[da['Survived']==1]['Age'],hist=False)
sns.distplot(da[da['Survived']==0]['Age'],hist=False)
```

## Heatmap (categorical-categorical)

```python
a=pd.crosstab(da['Pclass'],da['Survived'])
a
```

```python
sns.heatmap(a,annot=True,fmt='d')
```

```python
#to get perecentage of people survived based on the embarked
(da.groupby('Embarked').mean()['Survived']*100)
```

# clustermap - (categorical-categorical)

```python
sns.clustermap(pd.crosstab(da['SibSp'],da['Survived']))
```
