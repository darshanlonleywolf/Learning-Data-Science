```python
#should import request library to fetch the data from url
import requests
import pandas as pd
```

```python
requests.get('https://api.themoviedb.org/3/movie/popular?api_key=4b63febfa9089e4b4f50c1420e14a853&language=en-US&page=1')
```

```python
json_data=requests.get('https://api.themoviedb.org/3/movie/popular?api_key=4b63febfa9089e4b4f50c1420e14a853&language=en-US&page=1').json()
```

```python
json_data
```

```python
# Now to get the required data from the json_data object using loops and create a data frame
title=[]
original_language=[]
popularity=[]
vote_count=[]
overview=[]
release_date=[]


for i in json_data['results'] :
    popularity.append(i['popularity'])
    title.append(i['title'])
    original_language.append(i['original_language'])
    vote_count.append(i['vote_count'])
    release_date.append(i['release_date'])
    
d={'title':title, 'original_language':original_language, 'popularity':popularity,'vote_count':vote_count,'release_date':release_date}

df=pd.DataFrame(d)
df
```

```python
final=pd.DataFrame()
for j in range(1,31):
    json_data=requests.get('https://api.themoviedb.org/3/movie/popular?api_key=4b63febfa9089e4b4f50c1420e14a853&language=en-US&page={}'.format(j)).json()
    
    title=[]
    original_language=[]
    popularity=[]
    vote_count=[]
    overview=[]
    release_date=[]


    for i in json_data['results']:
        popularity.append(i['popularity'])
        title.append(i['title'])
        original_language.append(i['original_language'])
        vote_count.append(i['vote_count'])
        release_date.append(i['release_date'])

    d={'title':title, 'original_language':original_language, 'popularity':popularity,'vote_count':vote_count,'release_date':release_date}

    df=pd.DataFrame(d)
    final=final.append(df, ignore_index=True)
final.shape
```
