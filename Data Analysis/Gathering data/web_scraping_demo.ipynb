## `Webscraping of ambition box website, will gather the data of comanies, its rating, reviews, how many years old`


```python
import pandas as pd
import requests
from bs4 import BeautifulSoup #library for web scraping 
import numpy as np
```

```python
webpage=requests.get('https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav&page=1').text
print(webpage)
```

```python
bs=BeautifulSoup(webpage, 'lxml')
print(bs.prettify())
```

```python
comapany_name=[]
ratings=[]
no_of_reviews=[]
final=pd.DataFrame()

bs.find_all('h1')[0].text
bs
   
```

```python
for i in bs.find_all('h2'):
    print(i.text.strip())
```

```python
for i in bs.find_all('h2'):
    comapany_name.append(i.text.strip())
    
for i in bs.find_all('p',class_='rating'):
    ratings.append(i.text.strip())
    
for i in bs.find_all('a',class_='review-count'):
    no_of_reviews.append(i.text.strip())
    
a={'Company_Name':comapany_name,'ratings':ratings,'no_of_reviews':no_of_reviews}
final=pd.DataFrame(a)
final
```

```python
#differnt method based on the html code( extracting hole div and getting what we want from that)
full=bs.find_all('div', class_='company-content-wrapper')
len(full)
```

```python
name=[]
rating=[]
reviews=[]
sector=[]
old=[]
employees=[]

df=pd.DataFrame()


for i in full:
    name.append(i.find('h2').text.strip())
    rating.append(i.find_all('p', class_='rating')[0].text.strip())
    reviews.append(i.find('a',class_='review-count').text.strip())
    sector.append(i.find_all('p',class_='infoEntity')[0].text.strip())
    old.append(i.find_all('p',class_='infoEntity')[2].text.strip())
    try:
        employees.append(i.find_all('p',class_='infoEntity')[3].text.strip())
    except:
        employees.append(np.nan)
    
d={'Name':name,'rating':rating,'reviews':reviews,'sector':sector,'old':old,'employees':employees}
df=pd.DataFrame(d)
df  
    
```

```python
final=pd.DataFrame()

for j in range(0,31):
    webpage=requests.get('https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav&page={}'.format(j)).text

    bs=BeautifulSoup(webpage, 'lxml')
    full=bs.find_all('div', class_='company-content-wrapper')

    name=[]
    rating=[]
    reviews=[]
    sector=[]
    old=[]
    employees=[]

    df=pd.DataFrame()


    for i in full:
        name.append(i.find('h2').text.strip())
        rating.append(i.find_all('p', class_='rating')[0].text.strip())
        reviews.append(i.find('a',class_='review-count').text.strip())
        sector.append(i.find_all('p',class_='infoEntity')[0].text.strip())
        try:
            old.append(i.find_all('p',class_='infoEntity')[2].text.strip())
        except:
            old.append(np.nan)
        try:
            employees.append(i.find_all('p',class_='infoEntity')[3].text.strip())
        except:
            employees.append(np.nan)

    d={'Name':name,'rating':rating,'reviews':reviews,'sector':sector,'old':old,'employees':employees}
    df=pd.DataFrame(d)
    final=pd.concat((df,final),ignore_index=True)

display(final)

    
```

```python
final.to_csv('companies_with_ratings')
```
