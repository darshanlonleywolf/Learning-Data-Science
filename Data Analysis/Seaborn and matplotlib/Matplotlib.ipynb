## Matplotlib - visualization library used for data anlaysis

```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
#this is very important bcz whithout this the graphs will not be printing, to print the garphs ypu need this line
```

# `pie chart`

```python
areas=['marketing','development','hr','customer support']
budget=[2.5,8,2,5]
myexplode = [0.3, 0, 0, 0]
plt.pie(budget, labels=areas, explode = myexplode, autopct='%0.1f%%')
plt.legend()
plt.show()
```

```python
plt.pie(budget, labels=areas, explode = myexplode, autopct='%0.4f%%')
plt.show()
```

# `Scatter plot`

```python
x=[23,42,52,52,73,23] #random values 
y=[2,12,53,77,36,93]
plt.scatter(x,y, color='blue')
plt.xlabel('Age')
plt.ylabel('weight')
plt.title('Age/Weight')
```

# `2D plot`

```python
product_id=[1,2,3,4,5,6,7]
online_price=[23,52,25,77,36,27,85]
ofline=[2,5,63,34,85,14,24]

plt.plot(product_id,online_price, color='red', marker='+', markersize=11, label='online') #marker='d' dimaond shape at points
plt.plot(product_id,ofline, color='green', marker='+', markersize=11,label='ofline')
plt.title('product price comparision')
plt.legend()
plt.show()

```

# `Bar Graph`

```python
year=[2012,2013,2014,2015,2016,2017]
price=[21,42,51,51,51,45]
plt.bar(year,price,color='green')
plt.legend()
```
