## 1.Handling mixed data (numberical and categorical data in same column)

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('titanic.txt',usecols=['Ticket','Cabin','Fare','Survived'])
```

```python
df
```

```python
df['Cabin'].unique()
```

```python
#the below line extracts the numerical value
df['cabin_num']=df['Cabin'].str.extract('(\d+)')

#now will capture the first leter in the cabin column
df['cabin_cat']=df['Cabin'].str[0]

df.head()
```

```python
df.dropna()
```

# 2.Handling Date and Time variable (ex: 23-aug-2022...etc)


```python
time=pd.read_csv('https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/main/day34-handling-date-and-time/messages.csv')
date=pd.read_csv('https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/main/day34-handling-date-and-time/orders.csv')
```

```python
date.sample(5)
```

```python
time.sample(5)
```

`By default when you import data using pandas as a dataframe the date and time columns will be considered as object(string)
datatype, therefore we need to convert the datatype to date and time datatype`

```python
date.info()
```

```python
date['date']=pd.to_datetime(date['date'])
```

```python
date.info()
```

```python
# extract year
date['year']=date['date'].dt.year
date.sample(5)
```

```python
# extract month
date['month']=date['date'].dt.month
date.sample(5)
```

```python
# extract month names
date['month_name']=date['date'].dt.month_name()
date.sample(5)
```

```python
# extract date
date['date_day']=date['date'].dt.day
date.sample(5)
```

```python
# extract day name
date['day_of_the_week']=date['date'].dt.dayofweek
date.sample(5)
```

```python
# extract year
date['day_name']=date['date'].dt.day_name()
date.sample(5)
```

```python
# to check weeked
date['is_weekend']=np.where(date['day_name'].isin(['Sunday','Saturday']),1,0)
date.sample(5)
```
